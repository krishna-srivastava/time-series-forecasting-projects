```python
import pandas as pd
import joblib

# ===== LOAD MODEL & DATA =====
model   = joblib.load('../Model/randomforestmodel.pkl')
features = joblib.load('../Model/feature_columns.pkl')

df = pd.read_csv('../Data/features-column.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df = df.sort_index()

print(f"Model loaded!")
print(f"Data loaded: {df.shape}")
```

```python
# ===== LAST ROW SE PREDICT =====
last_row   = df[features].iloc[-1:]
last_date  = df.index[-1]
last_close = df['Close'].iloc[-1]

# Predict
direction   = model.predict(last_row)[0]
probability = model.predict_proba(last_row)[0]

# Next trading day
next_day = last_date + pd.Timedelta(days=1)
if next_day.weekday() == 5:
    next_day += pd.Timedelta(days=2)
elif next_day.weekday() == 6:
    next_day += pd.Timedelta(days=1)
```

```python
# ===== PRINT RESULT =====
print("\n" + "=" * 50)
print("   TCS — NEXT DAY PREDICTION")
print("=" * 50)
print(f"Last Date       : {last_date.date()}")
print(f"Last Close      : ₹{last_close:.2f}")
print(f"Prediction For  : {next_day.date()}")
print(f"\nDirection       : {'📈 UP' if direction==1 else '📉 DOWN'}")
print(f"UP Probability  : {probability[1]*100:.1f}%")
print(f"DOWN Probability: {probability[0]*100:.1f}%")
print(f"\nConfidence      : {'HIGH ' if max(probability)>0.6 else 'MEDIUM ' if max(probability)>0.55 else 'LOW '}")
print("=" * 50)
print("Educational purpose only. Not financial advice.")
```
