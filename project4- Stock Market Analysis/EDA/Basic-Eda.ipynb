```python
import pandas as pd
```

```python
# =====================================================
# Load Dataset & Basic Preprocessing
# =====================================================

df = pd.read_csv('../Data/TCS_stock_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head(5)
```

```python
print("Total Rows In The Dataset:",df.shape[0])
print("Total Columns In The Dataset:",df.shape[1],"\n")

print("Columns Names: ",df.columns,"\n")

print(f"Date Range: {df['Date'].min()} to {df['Date'].max()}\n")

total_days = (df['Date'].max() - df['Date'].min()).days
total_years = total_days / 365.25
print(f"Total Year: {total_years:.1f} Years (approx {total_days} days)")
```

```python
print("Missing Values In The Dataset:")
print(df.isnull().sum())
```

```python
print("Data Info:")
print(df.info())
```
