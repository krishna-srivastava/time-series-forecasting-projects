```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# =====================================================
# Load Dataset & Basic Preprocessing
# =====================================================

df = pd.read_csv('../Data/electricity_demand.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)   
```

```python
# Temperature vs Demand (scatter plot)
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='temperature' , y='demand_MW')
plt.title("Temperature vs Demand")
plt.show()
```

```python
# humidity vs Demand (scatter plot)
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='humidity' , y='demand_MW')
plt.title("humidity vs Demand")
plt.show()
```

```python
# wind_speed vs Demand (scatter plot)
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='wind_speed' , y='demand_MW')
plt.title("wind_speed vs Demand")
plt.show()
```

```python
# precipitation vs Demand (scatter plot)
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='precipitation' , y='demand_MW')
plt.title("precipitation vs Demand")
plt.show()
```

```python
# Correlation Heatmap
corr = df[['demand_MW', 'temperature', 'humidity', 'wind_speed', 'pressure', 'precipitation','holiday']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr,annot=True,fmt='.2f',cmap='coolwarm',center=0,linewidths=0.5,square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

print(corr)
```
